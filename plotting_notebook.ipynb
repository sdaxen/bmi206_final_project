```python
from __future__ import division
import os
import sys
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import cPickle as pkl
import glob
```

```python
def read_xl(f):
    '''Read cross-links from cross-link data file.'''
    xl_list = pkl.load(open(f, 'r'))
    xl_data = dict(map(lambda x: (tuple(sorted(("%s_%s" % (x[0][0], x[0][1]), "%s_%s" % (x[0][2], x[0][3])))), x[1]), xl_list))
    return xl_data

def read_xl_table(f):
    '''Read cross-link table and return set of cross-link pairs.'''
    xl_set = set([])
    with open(f, 'r') as fh:
        for line in fh:
            values = line.strip().split()
            pair = ("%s_%s" % (values[0], values[1]), "%s_%s" % (values[2], values[3]))
            xl_set.add(pair)
    return xl_set
```

```python
#retrieve and parse cross-link pickle files
xl_files = glob.glob("*comb*/scripts/kmeans_250_1/cluster.0/XL*.pkl")
xl_files.extend(glob.glob("*cv*/scripts/kmeans_250_1/cluster.0/XL*.pkl"))
xl_data = {f:read_xl(f) for f in xl_files}

#find out which cross-links are violated 
violated_xl = {x:set([k for k,v in y.items() if v<0]) for x,y in xl_data.items()}
n_xlinks = max([len(y.keys()) for x,y in xl_data.items()]) #number of cross-links
perc_satisfied = {k:1-len(v)/n_xlinks for k,v in violated_xl.items()} #percent of satisfied cross-links
#perc_satisfied
#all_xlinks_violated = [x for k,v in violated_xl.items() for x,y in v.items()]

#get cross-links set aside for testing
test_xl_files_edc = glob.glob("cross_validation_xlinks/EDC*test*")
test_xl_files_dss = glob.glob("cross_validation_xlinks/*DSS*test*")
test_xls = map(lambda x: read_xl_table(test_xl_files_edc[x]).union(read_xl_table(test_xl_files_dss[x])), range(len(test_xl_files_edc)))
```

```python
xl_labels = sorted(list({x for k,v in violated_xl.items() for x in v}), key=lambda x: (x[0].split('_')[0], x[1].split('_')[0], x[0].split('_')[1], x[1].split('_')[1]))
#print xl_labels
file_labels = ['All', 'CV1', 'CV2', 'CV3', 'CV4', 'CV5', 'CV6', 'CV7', 'CV8', 'CV9', 'CV10']
vals = np.array([[0. if x in violated_xl[f] else 1. for f in xl_files] for x in xl_labels]).T
test = [1-len([x for x in test_xls[i] if x in violated_xl[f]])/len(test_xls[i]) for i,f in enumerate(xl_files[1:])]
print "Cross-correlation score: %.4f" % (1-np.array(test)).mean()
#print dict([(f,[x for x in test_xls[i] if x in violated_xl[f]]) for i,f in enumerate(xl_files[1:])]).values()

#plot satisfaction (red) or violation (blue) of crosslinks for different runs
x_labels = ["%s - %s" % x for x in xl_labels]
fig, ax = plt.subplots(figsize=(14,8))
ax.imshow(vals, interpolation='nearest')
ax.yaxis.set_ticks(np.arange(11))
ax.set_yticklabels(file_labels)
ax.xaxis.set_ticks(np.arange(len(xl_labels)))
ax.set_xticklabels(x_labels, rotation=60)
plt.tight_layout()
```

```python
def read_precision(files):
    precision = {}
    subunits = set([])
    for f in files:
        with open(f, 'rU') as fh:
            for line in fh:
                entries = line.strip().split(' ')
                subunit = entries[0]
                if ' '.join(entries[3:6])!='average pairwise distance':
                    continue
                precision.setdefault(subunit, {})
                pair = (entries[1], entries[2])
                subunits.update(set(pair))
                precision[subunit][pair]=entries[-1]
    return precision, sorted(subunits)
                    
```

```python
precision_files = glob.glob("precision/*.out")
precision, subunits = read_precision(precision_files)
labels = ['All_1', 'All_2', 'Run1_1', 'Run1_2', 'Run2_1', 'Run2_2', 'Run3_1', 'Run3_2']

#Plot dRMSD precision for all
all_precision = precision["All"]
dmat = np.empty(shape=(len(labels), len(labels)))
for pair, drmsd in all_precision.items():
    coord = [subunits.index(pair[0]), subunits.index(pair[1])]
    dmat[coord[0],coord[1]]=drmsd
    dmat[coord[1],coord[0]]=drmsd

fig, ax = plt.subplots(figsize=(9,9))
cax = ax.imshow(dmat, interpolation='nearest', cmap=plt.cm.jet)
ax.yaxis.set_ticks(np.arange(len(labels)))
ax.set_yticklabels(labels, fontsize=13.5)
ax.xaxis.set_ticks(np.arange(len(labels)))
ax.set_xticklabels(labels, fontsize=13.5)
cbar = fig.colorbar(cax, orientation='vertical', shrink=0.725)
cbar.ax.tick_params(labelsize=15) 
#cbar.ax.set_xticklabels(['Low', 'Medium', 'High'])# horizontal colorbar
plt.tight_layout()
plt.grid(False)
plt.show()
```
